```python
# Maps
seasonsMap ={
    1: "Spring",
    2: "Summer",
    3: "Fall",
    4: "Winter"
}
weatherMap ={
    1: "Clear + Few Clouds + Partly cloudy + partly cloudy",
    2: "Mist + Couldy, Mist + Broken clouds, Mist + Few clouds, Mist",
    3: "Light Snow, Light Rain + Thunderstorm + Scattred clouds, Light Rain + scatter",
    4: "Heavy Rain + Ice pallets + Thunderstorm + Mist, snow + Fog"
}
```

```python
data['date'] =data.dteday.apply(lambda x : x.split()[0] )
data['weekday'] =data.date.apply(lambda dateString : calendar.day_name[datetime.strptime(dateString, "%Y-%m-%d").weekday()] )
data['season'] =data.season.map(seasonsMap)
data['weather'] =data.weathersit.map(weatherMap)
```

```python
categoryVarList =['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday','workingday']
for var in categoryVarList:
    data[var] =data[var].astype('category')
```

```python
data.head()
```

```python
data =data.drop(['dteday'], axis=1)
```

```python
# outlier detection
fig, axes =plt.subplots(nrows=4, ncols=1)
fig.set_size_inches(10, 20)
sns.boxplot(data=data, y='cnt', orient='v', ax=axes[0])
sns.boxplot(data=data, y='cnt', x='season', orient='v', ax=axes[1])
sns.boxplot(data=data, y='cnt', x='hr', orient='v', ax=axes[2])
sns.boxplot(data=data, y='cnt', x='workingday', orient='v', ax=axes[3])

axes[0].set(title='Count Distribuation', ylabel='Count')
axes[1].set(title='Count/Season Distribuation', xlabel='Season', ylabel='Count')
axes[2].set(title='Count/Hour Distribuation', xlabel='Hour of the day', ylabel='Count')
axes[3].set(title='Count/Season Distribuation', xlabel='Working Day', ylabel='Count')
```

```python
# Correlation Analysis
corrMat =data[['temp','atemp','casual','registered','hum','windspeed', 'cnt']].corr()
mask =np.array(corrMat)
mask[np.tril_indices_from(mask)] =False
fig, axes =plt.subplots()
fig.set_size_inches(15, 15)
sns.heatmap(corrMat, mask=mask, vmax=.8, square=True, annot=True )
```
