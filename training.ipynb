```python
data['date'] =data.dteday.apply(lambda x: x.split()[0] )
data['weekday'] =data.date.apply( lambda dateString : datetime.strptime(dateString, "%Y-%m-%d").weekday() )
```

```python
data.columns
```

```python
# features
categoricalFeatures =['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday','workingday']
numericalFeatures =['temp', 'atemp', 'hum', 'windspeed']
dropFeatures =['dteday','casual', 'registered', 'cnt', 'date']
```

```python
for var in categoricalFeatures:
    data[var] =data[var].astype('category')
```

```python
data.head(1)
```

```python
data =data[pd.notnull(data['cnt'])].sort_values(by=['dteday'])
y =data['cnt']
```

```python
data.head(1)
```

```python
data =data.drop(dropFeatures, axis=1)
```

```python
data.head(1)
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test =train_test_split(data, y, test_size=.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestRegressor

classifier =RandomForestRegressor(n_estimators=500)
classifier.fit(X=data, y=np.log1p(y))
```

```python
predictions =classifier.predict(X=X_test)
```

```python
predictions
```

```python
# Analyze the results
def rmsle(y, y_, convertExp =True):
    if convertExp:
        y =np.exp(y),
        y_ =np.exp(y_)
    log1 =np.nan_to_num(np.array([np.log(v + 1) for v in y ]))
    log2 =np.nan_to_num(np.array([np.log(v + 1) for v in y_ ]))
    calc =(log1 - log2) ** 2
    return np.sqrt(np.mean(calc))
```

```python
print("RMSLE",rmsle(np.exp(np.log1p(y_test)), np.exp(predictions), False))
```

```python
dataFrame =pd.DataFrame()
dataFrame['features'] =X_train.columns
dataFrame['coefficient'] =classifier.feature_importances_
dataFrame.sort_values(by=['coefficient'], ascending=False, inplace=True)
fig, axes =plt.subplots()
fig.set_size_inches(15, 5)
sns.barplot(data =dataFrame, x='features', y='coefficient', ax=axes)
```
